```python

import librosa
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import warnings
warnings.filterwarnings('ignore')  # Hides all pink warning boxes instantly

# A clean, simple memory dictionary to hold voice prints
voice_database = {}

def extract_voice_fingerprint(audio_path):
    """Extracts unique acoustic features from an audio file using pure math arrays."""
    # Load the audio data cleanly using standard floating points (skips torchaudio)
    y, sr = librosa.load(audio_path, sr=16000)

    # Extract Mel-Frequency Cepstral Coefficients (the acoustic footprint)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

    # Average the audio frequencies across the entire timeline to create a 1D feature vector
    vocal_vector = np.mean(mfccs.T, axis=0)
    return vocal_vector.reshape(1, -1)

print("🎉 Success! Mathematical audio feature engine is fully initialized.")
```

```python

#@title 🎙️ Speaker Verification Form { display-mode: "form" }
#@markdown Fill in the settings below to enroll voices or analyze matching logs.

action = "Verify Similarity" #@param ["Enroll Voice", "Verify Similarity"]
speaker_name = "Himangshu" #@param {type:"string"}
audio_file_path = "/content/Speaker1.wav" #@param {type:"string"}
compare_file_path = "/content/Speaker2.wav" #@param {type:"string"}

import os

if not os.path.exists(audio_file_path):
    print(f"❌ Error: Cannot find file '{audio_file_path}'. Open the 📁 folder icon to upload it first.")
else:
    if action == "Enroll Voice":
        print(f"🧬 Extracting unique vocal features for {speaker_name}...")
        vector = extract_voice_fingerprint(audio_file_path)
        voice_database[speaker_name] = vector
        print(f"✅ Successfully enrolled voice print for user: {speaker_name}!")

    elif action == "Verify Similarity":
        if not os.path.exists(compare_file_path):
            print(f"❌ Error: Comparison file '{compare_file_path}' is missing!")
        else:
            print(f"📊 Analyzing similarity between '{audio_file_path}' and '{compare_file_path}'...")
            vec1 = extract_voice_fingerprint(audio_file_path)
            vec2 = extract_voice_fingerprint(compare_file_path)

            # Calculate similarity using standard Cosine Distance math
            similarity_matrix = cosine_similarity(vec1, vec2)
            score = similarity_matrix[0][0]

            # Normalize negative mathematical arrays to clean percentages
            score_percent = max(0.0, (score + 1) / 2) * 100 if score < 0 else score * 100

            print("\n================ VERDICT REPORT ================")
            if score_percent >= 75.0:
                print("✅ MATCH VERDICT: Samples belong to the SAME speaker.")
            else:
                print("❌ MISMATCH VERDICT: Samples belong to DIFFERENT individuals.")
            print(f"📈 Confidence Match Matrix Score: {score_percent:.2f}%")
            print("==================================================")
```
